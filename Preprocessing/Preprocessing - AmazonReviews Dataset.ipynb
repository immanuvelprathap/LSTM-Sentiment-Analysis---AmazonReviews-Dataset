### Let us pre-process the data,

```python
import json
import pandas as pd
```

```python
data = []

for line in open('/Users/immanuvelprathaps/Desktop/LSTM/Cell_Phones_and_Accessories.json','r') :
    data.append(json.loads(line))
```

```python
print(data[0])
```

```python
df = pd.DataFrame(data)
```

```python
print(len(data))
```

```python
df.head(10)
```

```python
df = df.drop(columns=['reviewerName', 'vote','image','style'])
```

```python
df1 = df.rename(columns = {'overall': 'rating', 'asin': 'productID'}, inplace = False)
```

```python
df1.dropna(axis = 0, how ='any',inplace=True) 
```

```python
df1.drop_duplicates(subset=['rating','reviewText'],keep='first',inplace=True)
```

```python
print(len(df1))
```
